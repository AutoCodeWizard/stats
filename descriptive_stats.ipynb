## Übung 3: Varianz

### Aufgabe:
#### Berechnen Sie die Varianz der Zahlen. 
#### a) Verwenden Sie die Populationsvarianz 
#### b) Verwenden Sie die Stichprobenvarianz
```Python
ages: [21, 23, 25, 27, 29, 31, 33]  # Altersangaben
```
### Lösung a):
Die Populationsvarianz misst die durchschnittliche Abweichung jeder Zahl \
in der Population vom Populationsschwerpunkt (Mittelwert). \
Um die Populationsvarianz zu berechnen, verwenden wir die Formel:

Varianz = Σ(x - μ)² / N

Dabei ist μ der Mittelwert, x sind die einzelnen Zahlen und N ist die Anzahl der Zahlen.

Für die gegebenen Zahlen ergibt sich die Populationsvarianz:

Varianz = [(21-27)² + (23-27)² + (25-27)² + (27-27)² + (29-27)² + (31-27)² + (33-27)²] / 7 = 16.0

### Lösung b):
Die Stichprobenvarianz misst die durchschnittliche Abweichung jeder Zahl \
in der Stichprobe vom Stichprobenmittelwert. \
Die Berechnung erfolgt ähnlich wie bei der Populationsvarianz, \
aber der Teiler (N-1) wird verwendet, da es sich um eine Stichprobe handelt.

Varianz = Σ(x - x̄)² / (N-1)

Dabei ist x̄ der Stichprobenmittelwert, x sind die einzelnen Zahlen und N ist die Anzahl der Zahlen.

Für die gegebenen Zahlen ergibt sich die Stichprobenvarianz:

Varianz = [(21-27)² + (23-27)² + (25-27)² + (27-27)² + (29-27)² + (31-27)² + (33-27)²] / 6 ≈ 18.67

Die Populationsvarianz beträgt 16, während die Stichprobenvarianz etwa 18,67 beträgt.


````python
# definieren der Variable ages
ages = [21, 23, 25, 27, 29, 31, 33]

# a) Populationsvarianz
population_variance = np.var(ages)
print("Populationsvarianz:", population_variance)

# b) Stichprobenvarianz
sample_variance = np.var(ages, ddof=1) # Delta Degrees of Freedom 1 für Stichprobenvarianz
print("Stichprobenvarianz:", sample_variance)
````

**ergänzung zu b)**
Wir nehmen jetzt eine Stichprobe von 5 Altersangaben aus der ursprünglichen Gruppe:
```python
sample_ages = [21, 23, 27, 31, 33]
```
Wir wissen: Die Stichprobenvarianz ist die durchschnittliche Abweichung jeder Zahl in der Stichprobe vom Stichprobenmittelwert. \
Die Berechnung erfolgt ähnlich wie bei der Populationsvarianz, aber der Teiler (N-1) wird verwendet, da es sich um eine Stichprobe handelt.

Stichprobenmittelwert (x̄) = (21 + 23 + 27 + 31 + 33) / 5 = 27

Varianz = [(21-27)² + (23-27)² + (27-27)² + (31-27)² + (33-27)²] / 4 = 26

Die Stichprobenvarianz für diese Stichprobe beträgt nun:

````python
sample_ages = [21, 23, 27, 31, 33]

sample_variance = np.var(sample_ages, ddof=1) # Delta Degrees of Freedom 1 für Stichprobenvarianz
# ddof gibt an, wie viele Freiheitsgrade von der Stichprobe abgezogen werden sollen, 
# um eine konservative Schätzung der Varianz zu erhalten
print('Sample Variance:', sample_variance)
````

## Übung 4: Standardabweichung

### Aufgabe:
Gegeben ist `demo_list` aus Übung 1. \
Ihre Aufgabe ist es, die Standardabweichung dieser Zahlen zu berechnen.

### Lösung:
Die Standardabweichung ist ein Maß für die Streuung der Werte eines Datensatzes. \
Sie wird berechnet, indem man die Quadratwurzel der Varianz nimmt:

Standardabweichung = √Varianz

Dabei ist die Varianz der Durchschnitt der quadrierten Abweichungen vom Mittelwert:

Varianz = Σ(x - μ)² / N

Um die Standardabweichung zu berechnen, verwende erstmal die oben genannt Formel:

````python
# komplizierter Weg
def berechne_standardabweichung(numbers: list) -> float: 
    """
    Diese Funktion berechnet die Standardabweichung einer gegebenen Liste von Zahlen.
    
    Args:
    - numbers (list): Eine Liste von Zahlen, für die die Standardabweichung berechnet werden soll.
    
    Returns:
    - float: Die Standardabweichung der gegebenen Liste von Zahlen.
    """
    # Schritt 1: Mittelwert der Zahlen berechnen
    # Der Mittelwert gibt den Durchschnittswert der Daten an und wird häufig mit dem Symbol μ (Mü) bezeichnet.
    mittelwert = sum(numbers) / len(numbers)
    
    # Schritt 2: Quadratische Abweichungen vom Mittelwert berechnen
    # Die quadratische Abweichung ist die Quadratur der Differenz zwischen jeder Zahl und dem Mittelwert.
    # Das heißt, es ist das Quadrat des Abstands jeder Zahl zum Mittelwert.
    quad_abweichungen = [(x - mittelwert)**2 for x in numbers]
    
    # Schritt 3: Mittelwert der quadratischen Abweichungen berechnen, auch als Varianz bezeichnet
    # Die Varianz ist der Durchschnitt der quadratischen Abweichungen und wird mit dem Symbol σ² bezeichnet.
    varianz = sum(quad_abweichungen) / len(quad_abweichungen)
    
    # Schritt 4: Standardabweichung berechnen
    # Die Standardabweichung ist die Quadratwurzel der Varianz und wird mit dem Symbol σ (Sigma) bezeichnet.
    standardabweichung = varianz**0.5  
    
    # klar alternativ kann auch math.sqrt oder np.sqrt, np.mean.. ect. verwendet werden
    # aber hier sollte es erstmal gemacht werden, um zu verstehen, was dahinter steckt
    
    return standardabweichung


berechne_standardabweichung(demo_list)
````

## Übung 5: Quartile und Perzentile

### Aufgabe:
Gegeben ist der Array `numbers` aus Übung 1. Ihre Aufgabe ist es, das erste Quartil (25. Perzentil), das zweite Quartil (50. Perzentil oder Median) und das dritte Quartil (75. Perzentil) dieser Zahlen zu berechnen.

### Lösung:
Quartile sind spezielle Punkte in einem Datensatz, die die Daten in gleich große Abschnitte teilen. In diesem Fall teilen sie den Datensatz in vier Teile:

- **Erstes Quartil (25. Perzentil)**: Der Wert, unterhalb dessen 25% der Datenpunkte liegen.
- **Zweites Quartil (50. Perzentil oder Median)**: Der Wert, der den Datensatz in zwei Hälften teilt. 50% der Datenpunkte liegen unterhalb und 50% oberhalb dieses Wertes.
- **Drittes Quartil (75. Perzentil)**: Der Wert, unterhalb dessen 75% der Datenpunkte liegen.

Um diese Quartile zu berechnen, können Sie die Funktion `np.percentile` aus der NumPy-Bibliothek verwenden:

```python
first_quartile = np.percentile(numbers, 25)
second_quartile = np.percentile(numbers, 50)
median = np.median(numbers)  # Identisch mit dem 50. Perzentil
third_quartile = np.percentile(numbers, 75)

````python
# erstmal wieder komplizierter Weg ohne numpy
def berechne_quartile(numbers: list) -> tuple:
    """
    Diese Funktion berechnet die ersten, zweiten und dritten Quartile einer gegebenen Liste von Zahlen.
    
    Args:
    - numbers (list): Eine Liste von Zahlen, für die die Quartile berechnet werden sollen.
    
    Returns:
    - tuple: Ein Tuple, das den ersten, zweiten und dritten Quartil der gegebenen Liste von Zahlen enthält.
    """
    # Schritt 1: Liste sortieren
    # Ein sortiertes Array ist notwendig, um die Quartile zu finden.
    sorted_numbers = sorted(numbers)
    
    # Schritt 2: Länge der Liste ermitteln
    n = len(sorted_numbers)
    
    # Schritt 3: Quartile berechnen
    # Der k-te Quartil ist das k*25%-te Element im sortierten Array (0-basiert).
    def find_quartile(p):
        index = int(p * (n - 1)) # p * (n - 1) ist der Index des k-ten Quartils
        if n % 2 == 0: # Wenn die Anzahl der Elemente gerade ist, dann ist der Median der Durchschnitt der beiden mittleren Elemente
            return (sorted_numbers[index] + sorted_numbers[index + 1]) / 2 # index + 1 ist der Index des nächsten Elements
        else: # Wenn die Anzahl der Elemente ungerade ist, dann ist der Median das mittlere Element
            return sorted_numbers[index] #  index ist der Index des mittleren Elements
    
    first_quartile = find_quartile(.25)
    second_quartile = find_quartile(.5)  # Das ist auch das Median
    third_quartile = find_quartile(.75)
    
    return first_quartile, second_quartile, third_quartile

# Verwendung der Funktion
first_quartile, second_quartile, third_quartile = berechne_quartile(demo_list)
print("Erstes Quartil:", first_quartile)
print("Zweites Quartil (Median):", second_quartile)
print("Drittes Quartil:", third_quartile)
````

## Unterschied zwischen Quartilen und Perzentilen

Beide Begriffe, Quartile und Perzentile, beziehen sich auf statistische Maßzahlen zur Beschreibung der Verteilung eines Datensatzes. Sie unterscheiden sich jedoch in der Art und Weise, wie sie den Datensatz aufteilen:

### Quartile

Quartile teilen einen sortierten Datensatz in **vier gleiche Teile**. Dadurch entstehen drei Quartilwerte:

- **Erstes Quartil (Q1)**: Ist der Median der unteren Hälfte der Daten und trennt die untersten 25% der Daten von den obersten 75%.
- **Zweites Quartil (Q2)**: Ist der Median des gesamten Datensatzes und trennt die untersten 50% der Daten von den obersten 50%.
- **Drittes Quartil (Q3)**: Ist der Median der oberen Hälfte der Daten und trennt die untersten 75% der Daten von den obersten 25%.

### Perzentile

Perzentile sind allgemeiner und können einen sortierten Datensatz in **100 gleiche Teile** teilen, wodurch 99 Perzentilwerte entstehen.

- **p-te Perzentil**: Ist der Wert, unterhalb dessen p% der Beobachtungen fallen.

Zum Beispiel:

- Das 25. Perzentil entspricht dem ersten Quartil.
- Das 50. Perzentil entspricht dem zweiten Quartil (dem Median).
- Das 75. Perzentil entspricht dem dritten Quartil.

### Fazit

Kurz gesagt, Quartile sind spezielle Fälle von Perzentilen:

- **Das erste Quartil entspricht dem 25. Perzentil.**
- **Das zweite Quartil (der Median) entspricht dem 50. Perzentil.**
- **Das dritte Quartil entspricht dem 75. Perzentil.**

Da Perzentile feiner unterteilt sind, bieten sie eine detailliertere Beschreibung der Verteilung eines Datensatzes.


## Übung 6: Diagrammtypen

### Aufgabe 6A:
Erstellen Sie eine 2x2 Subfigure-Matrix in Python, in der folgende Diagramme dargestellt werden: 
- Balkendiagramm
- Histogramm
- Liniendiagramm
- Tortendiagramm

### Lösung 6A:

````python
# Daten für Aufgabe 6A
x = np.arange(0., 10., 0.2)
y1 = np.sin(x)
y2 = np.cos(x)
categories = ['A', 'B', 'C', 'D', 'E']
values = [7, 15, 5, 20, 10]
hist_data = np.random.randn(100)

# 2x2 Subplots für Aufgabe 6A
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Balkendiagramm
axs[0, 0].bar(categories, values)
axs[0, 0].set_title('Balkendiagramm')

# Histogramm
axs[0, 1].hist(hist_data, bins=20, color='green', alpha=0.7)
axs[0, 1].set_title('Histogramm')

# Liniendiagramm
axs[1, 0].plot(x, y1, label='sin(x)')
axs[1, 0].plot(x, y2, label='cos(x)', linestyle='--')
axs[1, 0].set_title('Liniendiagramm')
axs[1, 0].legend()

# Tortendiagramm
axs[1, 1].pie(values, labels=categories, autopct='%1.1f%%')
axs[1, 1].set_title('Tortendiagramm')

# Abstand zwischen den Plots anpassen
plt.tight_layout()
plt.show()
````

### Aufgabe 6B:
Erstellen Sie eine 2x2 Subfigure-Matrix in Python, in der folgende Diagramme dargestellt werden:
- Streudiagramm (Scatter Plot)
- Boxplot
- Heatmap
- Blasendiagramm (Bubble Chart)

### Lösung 6B:

````python
# Daten für Aufgabe 6B
np.random.seed(0)  # Für Reproduzierbarkeit
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.randn(100)
sizes = np.random.rand(50) * 1000  # Für das Blasendiagramm
matrix = np.random.rand(10, 12)  # Für die Heatmap

# 2x2 Subplots für Aufgabe 6B
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Streudiagramm
axs[0, 0].scatter(x, y)
axs[0, 0].set_title('Streudiagramm')
axs[0, 0].grid(True)

# Boxplot
axs[0, 1].boxplot(z)
axs[0, 1].set_title('Boxplot')

# Heatmap
sns.heatmap(matrix, ax=axs[1, 0], cmap='viridis')
axs[1, 0].set_title('Heatmap')

# Blasendiagramm
axs[1, 1].scatter(x, y, s=sizes, alpha=0.5)
axs[1, 1].set_title('Blasendiagramm')
axs[1, 1].grid(True)

# Layout anpassen
fig.tight_layout()
plt.show()
````
